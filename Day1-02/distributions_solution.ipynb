# Probability distributions: a quick refresher

The following example and exercise are aimed to refresh your mind on some of the key statistical distributions. You will use [SciPy statistical function](https://docs.scipy.org/doc/scipy/reference/stats.html) module and work with several probability distributions through the following steps: 

- Generating the mean, variance, skewness, and kurtosis of each distribution
- Drawing a sample from each distribution and plotting it

Let's start by importing the necessary libraries.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, beta, bernoulli, norm, t, gamma, expon, binom
```

## Example: Normal distribution
> `norm` function

```python
# set the mean and standard deviation
m, s = 0, 1 

# find the theoretical properties
mean, var, skew, kurt = norm.stats(m, s, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from normal(m,s) distribution 
sample = norm.rvs(m, s, size = (10000,1)) 

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Normal PDF')
plt.show()
```

### Students' T-distribution
> `t` function

```python
# set the degrees of freedom (d>0)
d = 5

# find the theoretical properties
mean, var, skew, kurt = t.stats(d, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from Students' T-distribution
sample = t.rvs(d, size = (10000,1)) 

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Students-T PDF')
plt.show()
```

### Uniform distribution
> `uniform` function

```python
# set the boundaries (-inf < a < b < inf)
a, b = 0, 1

# find the theoretical properties
mean, var, skew, kurt = uniform.stats(a, b, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from uniform(a,b) distribution 
sample = uniform.rvs(a, b, size = (10000,1)) 

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Uniform PDF')
plt.show()
```

### Exponential distribution
> `expon` function

```python
# set the scale parameter (s>0)
s = 5

# find the theoretical properties
mean, var, skew, kurt = expon.stats(loc=0, scale=s, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from Exponential(s) distribution 
sample = expon.rvs(loc=0, scale=s, size = (10000,1)) 

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Exponential PDF')
plt.show()
```

### Gamma distribution
`gamma` function

```python
# set the shape and scale parameters (a>0, s>0)
a, s = 1, 5

# find the theoretical properties
mean, var, skew, kurt = gamma.stats(a, loc=0, scale=s, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from gamma(a,s) distribution 
sample = gamma.rvs(a, loc=0, scale=s, size = (10000,1)) 

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Gamma PDF')
plt.show()
```

### Bernoulli distribution
> `bernoulli` function

```python
# set the probability of success (0<=p<=1)
p = 0.5

# find the theoretical properties
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from bernoulli(p) distribution 
sample = bernoulli.rvs(p,size = (10000,1))

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Bernoulli PMF')
plt.show()
```

### Binomial distribution
> `binom` function

```python
# set the number of trials and probability of success (n>0, 0<=p<=1)
n, p = 1, 0.5

# find the theoretical properties
mean, var, skew, kurt = binom.stats(n,p, moments='mvsk')
display(mean, var, skew, kurt)
```

```python
# drawing 10000 samples from Binomial(n,p) distribution 
sample = binom.rvs(n,p,size = (10000,1))

# plot probability distribution
plt.hist(sample, bins='auto', density=True)
plt.xlabel('sample values')
plt.title('Binomial PMF')
plt.show()
```
